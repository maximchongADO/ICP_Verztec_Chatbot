```python
def get_all_file_paths(directory, extensions=[".pdf", ".doc", ".docx",".txt"]):
    file_paths = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            # Filter files by the given extensions (e.g., PDF, DOC, DOCX files)
            if any(file.lower().endswith(ext) for ext in extensions):
                absolute_path = os.path.abspath(os.path.join(root, file))
                file_paths.append(absolute_path)
    return file_paths
cleaned_files = get_all_file_paths(r"C:\Poly\Poly Y3S1\Projects\ICP\Bot_reponse_data\cleaned")
```

```python
fallback_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separators=["\n\n", "\n", ".", "!", "?"]
)

def split_by_bullets_consume(text: str, max_chunk_size: int = 600):
    """
    Split text into chunks, prioritizing bullet point structure.
    - Detects lines starting with common bullet markers (numeric, alphabetic, or symbols).
    - Groups each bullet with its subsequent non-bullet lines as one chunk.
    - Falls back to paragraph splitting if no bullet points are present.
    - Ensures no content is skipped, and splits large chunks into sub-chunks.
    """
    # Regex for bullet start (ordered or unordered markers).
    bullet_pattern = re.compile(
        r'^[ \t]*(?:[\-\*•]|(?:\d+|[A-Za-z])[.)])\s+'
    )
    
    lines = text.splitlines()
    chunks = []
    current_chunk_lines = []
    in_bullet_mode = False
    
    for line in lines:
        if bullet_pattern.match(line):
            # This line is a bullet point (start of a new bullet chunk).
            if in_bullet_mode:
                # End the previous bullet chunk before starting a new one.
                if current_chunk_lines:
                    chunk_text = " ".join(part.strip() for part in current_chunk_lines if part.strip())
                    if chunk_text:
                        chunks.append(chunk_text)
                current_chunk_lines = [line.strip()]
            else:
                # If coming from normal text mode, flush any accumulated text chunk.
                if current_chunk_lines:
                    chunk_text = " ".join(part.strip() for part in current_chunk_lines if part.strip())
                    if chunk_text:
                        chunks.append(chunk_text)
                # Start bullet mode
                in_bullet_mode = True
                current_chunk_lines = [line.strip()]
        else:
            # This line does not start with a bullet.
            if in_bullet_mode:
                if line.strip() == "":
                    # Blank line indicates end of the bullet list or a break in content.
                    if current_chunk_lines:
                        chunk_text = " ".join(part.strip() for part in current_chunk_lines if part.strip())
                        if chunk_text:
                            chunks.append(chunk_text)
                    # Reset state to handle subsequent text outside bullet list
                    current_chunk_lines = []
                    in_bullet_mode = False
                else:
                    # Continuation of the current bullet item (same chunk).
                    current_chunk_lines.append(line.strip())
            else:
                # Normal text mode (no bullet list active).
                if line.strip() == "":
                    # Paragraph break in normal text.
                    if current_chunk_lines:
                        chunk_text = " ".join(part.strip() for part in current_chunk_lines if part.strip())
                        if chunk_text:
                            chunks.append(chunk_text)
                        current_chunk_lines = []
                else:
                    # Accumulate lines for a paragraph chunk.
                    current_chunk_lines.append(line.strip())
    
    # Flush any remaining lines in current_chunk_lines as the last chunk
    if current_chunk_lines:
        chunk_text = " ".join(part.strip() for part in current_chunk_lines if part.strip())
        if chunk_text:
            chunks.append(chunk_text)
    
    
    
    # Post-process chunks: split any chunk that exceeds max_chunk_size into smaller sub-chunks.
    final_chunks = []
    for chunk in chunks:
        if len(chunk) <= max_chunk_size:
            final_chunks.append(chunk)
        else:
            # Split large chunk into sentences or smaller pieces without losing context.
            # We attempt to split by sentence boundaries (". ", "? ", "! ") for coherence.
            sentences = re.split(r'(?<=[.!?])\s+', chunk)
            sub_chunk = ""
            for sent in sentences:
                if not sent:
                    continue
                if len(sub_chunk) + len(sent) + 1 <= max_chunk_size:
                    # Add sentence to current sub_chunk (plus a space if needed).
                    sub_chunk = sent if not sub_chunk else sub_chunk + " " + sent
                else:
                    # Current sub_chunk is full, store it and start a new one.
                    if sub_chunk:
                        final_chunks.append(sub_chunk.strip())
                    sub_chunk = sent  # start new sub-chunk with the current sentence
                    # If even a single sentence is longer than max_chunk_size, break it up raw
                    while len(sub_chunk) > max_chunk_size:
                        final_chunks.append(sub_chunk[:max_chunk_size])
                        sub_chunk = sub_chunk[max_chunk_size:]
            # Append the last accumulated sub_chunk
            if sub_chunk:
                final_chunks.append(sub_chunk.strip())
    print(final_chunks)
    return final_chunks





all_chunks_smarter2=[]
for file_path in cleaned_files:
    print(file_path)
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()

        smart_chunks = split_by_bullets_consume(text.lower())

        docs = [
        langDocument(
            page_content=chunk,
            metadata={
                "source": os.path.splitext(os.path.basename(file_path))[0],
                "doc_type": "text",
                "chunk_id": f"{os.path.splitext(os.path.basename(file_path))[0]}_{i}"
            }
        )
        
        for i, chunk in enumerate(smart_chunks)
    
    ]
    all_chunks_smarter2.extend(docs)

    
```

```python
import spacy

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

def chunk_with_spacy(text, max_tokens=100, max_chars=600):
    """
    Split text into semantically coherent chunks using spaCy sentence boundaries.
    Combines consecutive sentences into chunks of limited size (in tokens or characters).
    """
    doc = nlp(text)
    chunks = []
    current_chunk = ""
    current_tokens = 0

    for sent in doc.sents:
        sent_text = sent.text.strip()
        sent_tokens = len(sent)

        # If adding this sentence exceeds limits, start a new chunk
        if (current_tokens + sent_tokens > max_tokens) or (len(current_chunk) + len(sent_text) > max_chars):
            if current_chunk.strip():
                chunks.append(current_chunk.strip())
            current_chunk = sent_text
            current_tokens = sent_tokens
        else:
            current_chunk += " " + sent_text
            current_tokens += sent_tokens

    # Add final chunk
    if current_chunk.strip():
        chunks.append(current_chunk.strip())

    return chunks

all_chunks_smarter2=[]
for file_path in cleaned_files:
    print(file_path)
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read()

        smart_chunks = chunk_with_spacy(text.lower())

        docs = [
        langDocument(
            page_content=chunk,
            metadata={
                "source": os.path.splitext(os.path.basename(file_path))[0],
                "doc_type": "text",
                "chunk_id": f"{os.path.splitext(os.path.basename(file_path))[0]}_{i}"
            }
        )
        
        for i, chunk in enumerate(smart_chunks)
    
    ]
    all_chunks_smarter2.extend(docs)


```
